# Assignment: The cervical cancer data set

Download the cervical cancer data set provided, import it using `read_csv`.

- How many rows and columns are there?
- How many columns contain floating point numbers (float64)?
- How many of the subjects are smokers?
- Calculate the percentage of smokers
- Plot the age distribution (with e.g. 50 bins)
- Get the mean and standard distribution of age of first sexual intercourse





```python
from pandas import read_csv
df_cervical_cancer = read_csv('data/cervical_cancer.csv')
df_cc = df_cervical_cancer
df_cc
```

```python
no_rows = len(df_cc)
no_cols = len(df_cc.columns)
print('number of rows: ', no_rows, 'number of columns: ', no_cols)
```

```python
n = 0
for coltype in df_cc.dtypes:
    if coltype == 'float64':
        n += 1
print ('There are ', n, 'columns contain floating point numbers')
```

```python
n_smoker = 0
smoker_df_cc = df_cc['Smokes']
for smoker in smoker_df_cc:
    if smoker == 1.0:
        n_smoker += 1

print('Number of smoker is ', n_smoker)
```

```python
total_n = len(df_cc['Smokes'])
perc_smoker = (n_smoker/total_n)*100
perc_smoker = round(perc_smoker, 3)
print('The percentage of smoker is ', perc_smoker, '%')
```

```python
from matplotlib.pyplot import subplots, show
fig, ax = subplots()
ax.hist(df_cc['Age'], bins = 50) 
ax.set(xlabel = 'Age', ylabel = 'Count')
ax.set_title('Age Distribution')
show()
```

```python
df_cc_desc = df_cc['First sexual intercourse'].describe()
print ('mean of age of first sexual intercourse is ', round(df_cc_desc['mean'], 2))
print ('standard distribution of age of first sexual intercourse is ', round(df_cc_desc['std'], 2))
```
